```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("A1.csv")
df
```

```python
df['Classes'].value_counts()
```

# c

```python
df2 = df[df['Classes'] == 'A']
df3 = df[df['Classes'] == 'B']
df4 = df[df['Classes'] == 'C']

plt.hist(df2['Feature 1'])
```

# d

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.boxplot([df2['Feature 2'], df3['Feature 2'], df4['Feature 2']]) 
ax.set_xticklabels(['Class A','Class B','Class C'])
ax.set_ylabel('Frequency')
plt.show()
```

# e

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.violinplot([df2['Feature 3'], df3['Feature 3'], df4['Feature 3']])

ax.set_ylabel('Frequency')
plt.show()
```

# f

```python
plt.figure(figsize=(12,6))
plt.scatter(df2['Feature 3'], df2['Feature 1'])
plt.title('Class A')
plt.xlabel('Feature 3')
plt.ylabel('Feature 1')
plt.legend()
plt.show()

plt.figure(figsize=(12,6))
plt.scatter(df3['Feature 3'], df3['Feature 1'])
plt.title('Class B')
plt.xlabel('Feature 3')
plt.ylabel('Feature 1')
plt.legend()
plt.show()

plt.figure(figsize=(12,6))
plt.scatter(df4['Feature 3'], df4['Feature 1'])
plt.title('Class C')
plt.xlabel('Feature 3')
plt.ylabel('Feature 1')
plt.legend()
plt.show()
```

# G

```python
plt.figure(figsize=(12,6))
x = np.linspace(df2['Feature 4'].min(), df2['Feature 4'].max(), 100)
y = np.linspace(df2['Feature 1'].min(), df2['Feature 1'].max(), 100)
X, Y = np.meshgrid(x, y)
Z = (X * 0.001) + (Y * 0.1) + np.random.randn(*X.shape)  
cp = plt.contour(X, Y, Z)
plt.title('Class A')
plt.xlabel('Feature 4')
plt.ylabel('Feature 1')
plt.colorbar(cp)
plt.show()


plt.figure(figsize=(12,6))
x = np.linspace(df3['Feature 4'].min(), df3['Feature 4'].max(), 100)
y = np.linspace(df3['Feature 1'].min(), df3['Feature 1'].max(), 100)
X, Y = np.meshgrid(x, y)
Z = (X * 0.001) + (Y * 0.1) + np.random.randn(*X.shape)  
cp = plt.contour(X, Y, Z)
plt.title('Class B')
plt.xlabel('Feature 4')
plt.ylabel('Feature 1')
plt.colorbar(cp)
plt.show()

plt.figure(figsize=(12,6))
x = np.linspace(df4['Feature 4'].min(), df4['Feature 4'].max(), 100)
y = np.linspace(df4['Feature 1'].min(), df4['Feature 1'].max(), 100)
X, Y = np.meshgrid(x, y)
Z = (X * 0.001) + (Y * 0.1) + np.random.randn(*X.shape)  
cp = plt.contour(X, Y, Z)
plt.title('Class C')
plt.xlabel('Feature 4')
plt.ylabel('Feature 1')
plt.colorbar(cp)
plt.show()
```

# H

```python
plt.figure ( figsize= (12,6))
plt.hexbin(df2['Feature 2'], df2['Feature 4'])
plt.title('Hexagonal Bin Plot')
plt.xlabel('Feature 2')
plt.ylabel('Feature 4')
plt.colorbar(label='Density')
plt.show()
```

# I

```python
import seaborn as sns
df5 = df.drop(columns = ['Sample Number', 'Classes', 'Sex'])
corr_matrix = df5.corr()

plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
