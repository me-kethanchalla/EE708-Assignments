```python
stores = df['store'].unique()
stores = stores.tolist()

products = df['product'].unique()
products= products.tolist()
```

# c

```python
cust = df['name'].unique()
cust = cust.tolist()
number_of_obs = [0]*len(cust)
sales = [0]*len(cust)
df3 = pd.DataFrame( {'number_of_obs' : number_of_obs, 'sales' : sales} , index=cust)

for index in range(len(df)) :
    cus = df.iloc[index]['name'] 
    sal = df.iloc[index]['sales']
    df3.at[cus, 'number_of_obs'] += 1
    df3.at[cus,'sales'] += sal
 
  
df3
```

```python

number_of_obs = [0]*len(stores)
sales = [0]*len(stores)
df4 = pd.DataFrame( {'number_of_obs' : number_of_obs, 'sales_mean' : sales} , index=stores)

for index in range(len(df)) :
    sto = df.iloc[index]['store'] 
    sal = df.iloc[index]['sales']
    df4.at[sto, 'number_of_obs'] += 1
    df4.at[sto,'sales_mean'] += sal

df4['sales_mean'] = df4['sales_mean'] / df4['number_of_obs']
df4
```
