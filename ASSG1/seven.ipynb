```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
data1 = [1.75, 1.92, 2.62, 2.35, 3.09, 3.15,
2.53, 1.91]
data2 = [ 1.83, 1.99, 3.13, 3.29, 2.65, 2.87, 3.40, 2.46, 1.89, 3.35]
```

```python
def calculate (data) :

    mean = sum(data) / len(data)
    variance = 0 
    for x in range(len(data)) :
        variance += (data[x] - mean) ** 2  / (len(data)-1)  
    std_deviation = math.sqrt(variance)

    print("Mean:", mean)
    print("Standard Deviation:", std_deviation)
```

```python
calculate (data1)
calculate(data2)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.boxplot([data1, data2])
ax.set_xticklabels(['Data1', 'Data2'])
ax.set_ylabel('Values')

plt.tight_layout()
fig.savefig('seven.jpg')
plt.show()
```
