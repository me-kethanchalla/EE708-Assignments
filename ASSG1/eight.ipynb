```python
import numpy as np
import pandas as pd
```

```python
data = {
'Name': ['P. Lee', 'R. Jones', 'J. Smith', 'A. Patel', 'M. Owen', 'S. Green', 'N. Cook', 'W. Hands', 'P. Rice', 'F. Marsh'],
'Weight': [50, 115, 96, 41, 79, 109, 73, 104, 64, 136],
'Height': [1.52, 1.77, 1.83, 1.55, 1.82, 1.89, 1.76, 1.71, 1.74, 1.78],
'SBP': [68, 110, 88, 76, 65, 114, 108, 107, 101, 121],
'DBP': [112, 154, 136, 125, 105, 159, 136, 145, 132, 165],
'Diabetes': [0, 1, 0, 0, 0, 1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)
df
```

```python
def normalize(values):
    min_value = min(values)
    max_value = max(values)
    return [(x - min_value) / (max_value - min_value) for x in values]

df['normalised weight'] = normalize(df['Weight'])
df.to
```

```python
for index in range(len(df)):
    if df.iloc[index]['Weight'] < 60:
        df.at[index, 'binning'] = 'low'
        
    elif 60 <= df.iloc[index]['Weight'] <= 100:
        df.at[index, 'binning'] = 'medium'
    
    else:
        df.at[index, 'binning'] = 'high'


df        
```
